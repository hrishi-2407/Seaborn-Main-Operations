```python
'''
=================== Categorical Plots ====================

Count Plot (sns.countplot) -->
  Shows the count of observations in each categorical bin using bars.
  Example: sns.countplot(x='category', data=data)

Bar Plot (sns.barplot) -->
  Displays the mean (or other aggregate measure) of a quantitative variable for each level of a categorical variable.
  Example: sns.barplot(x='category', y='value', data=data)

Box Plot (sns.boxplot) -->
  Shows the distribution of a quantitative variable across different categories.
  Example: sns.boxplot(x='category', y='value', data=data)

Strip Plot (sns.stripplot) -->
  Draws a scatter plot where one variable is categorical.
  Example: sns.stripplot(x='category', y='value', data=data)

Point Plot (sns.pointplot) -->
  Shows point estimates and confidence intervals using scatter plot glyphs.
  Example: sns.pointplot(x='category', y='value', data=data)


===================== Regression Plots ======================

Linear Regression Plot (sns.lmplot) -->
  Fits and plots a linear regression model.
  Example: sns.lmplot(x='x_variable', y='y_variable', data=data)

Joint Plot (sns.jointplot) -->
  Combines univariate and bivariate plots to show the relationship between two variables along with their marginal distributions.
  Example: sns.jointplot(x='x_variable', y='y_variable', data=data, kind='reg')

Pair Plot (sns.pairplot) -->
  Creates a grid of plots to show relationships between pairs of variables.
  Example: sns.pairplot(data)

Heatmap (sns.heatmap) -->
  Shows a matrix of values as colors; while not strictly a regression plot, it can be used to display correlations.
  Example: sns.heatmap(data.corr(), annot=True)
'''
```
